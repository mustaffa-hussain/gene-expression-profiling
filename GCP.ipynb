## part 2
Questions	
1.	What does the analysis say about the general behavior of the different samples?
2.	Are the neuroendocrine tumors clearly separable from the adenocarcinoma tumors?
3.	What can be said about the variance of the PCA?


```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(my_data.transpose())
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
col_meta_data['histological_type_other'].value_counts()
```

```python
neuro_type=['neuroendocrine carcinoma nos','neuroendocrine','82463 neuroendocrine carcinoma nos','neuroendocrine carcinoma']
        
A=col_meta_data['histological_type_other'].values

for i, ele in enumerate(A):
    if ele in (neuro_type):
        A[i]='neuroendocrine'

principalDf['histological_type_other']=A
principalDf['histological_type_other'].value_counts()
```

only 8 sample is belongs to neuroendocrine type.

```python
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.scatterplot(x="principal component 1", y="principal component 2", data=principalDf, hue="histological_type_other",s=80)
```

#### From the above plot is is very much evedent that they belong to separe clusters and are easily differentiated. They fall in the top left corner of the plot, very different from rest of the samples.

```python
pca.explained_variance_ratio_

print('The amount of varience explained by first 2 components are : ',sum(pca.explained_variance_ratio_)*100)
```

#### we can observe that only 31% of varience is accounted for in the 2 major principal components. It is intelligent to suggest usage of more axes to preserve more information or use other method of dimensionality reduction. only for plotting purpose t-sne can be used.
